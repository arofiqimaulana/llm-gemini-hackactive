```python
# models/gemini-2.5-flash
# models/gemma-3-27b-it

PROMPT = """
Make a itinerary of 1 day vacation in bali.

Output in table, please include price and budget.

Make sure the plan is as cheap as possible and include beach and seafood.
And include the most famous bali spot.

Please make a plan how you will for research and planning of the itinerary first.
And then give the table.

Please create 3 plans first, and then evaluate and choose the cheapest plan.
"""

response = client.models.generate_content(
    model="models/gemini-2.5-flash",
    contents=[
        PROMPT,
    ]
)
```

```python
from IPython.display import Markdown

Markdown(response.text)
```

```python
PROMPT = """
Make a plan of breakfast, lunch, dan dinner in Bali with budget around Rp 200k - 300k.

Below is the relevant context:
{context}
"""

# 1) Baca file context
with open("bali.txt", "r", encoding="utf-8") as f:
    context = f.read()
```

```python
# 2) Format prompt
final_prompt = PROMPT.format(context=context)
```

```python
# 3) Generate
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=final_prompt,
)
```

## Configuration

```python
# models/gemini-2.5-flash
# models/gemma-3-27b-it

from google.genai import types

chat_config = types.GenerateContentConfig(
    # max_output_tokens=3000,
    temperature=0,
    # top_p=0.9,
    top_k=1,
)

PROMPT = """
What is LLM?
"""

response = client.models.generate_content(
    model="models/gemini-2.5-flash",
    contents=[
        PROMPT,
    ],
    config=chat_config
)

Markdown(response.text)
```
